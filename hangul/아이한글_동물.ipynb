```python
from bs4 import BeautifulSoup
import csv
import requests
import pandas as pd
import re
from urllib.request import urlopen
```

```python
url = "http://www.ihangul.kr/main/bbs/board.php?bo_table=m2_1a&sca=%EB%8F%99%EB%AC%BC"
```

```python
response = requests.get(url)
response
```

```python
soup = BeautifulSoup(urlopen(url),"html.parser")
```

```python
soup
```

```python
link_list = []

for link in soup.findAll('h2'):
    link_list.append(link.get_text().strip())
```

```python
link_list
```

```python
i = 1
temp_list = []
for i in range(0,len(link_list)):
    temp_list.append(link_list[i].split()[0])
print(temp_list)
```

```python
file = open('아이한글_동물.txt', 'w', encoding ='utf-8')

for temp in temp_list :
    file.write (temp + '\n')

file.close()
```
