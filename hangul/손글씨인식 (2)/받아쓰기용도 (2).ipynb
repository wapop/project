```python
백설공주파일 = open('./data/백설공주.txt',encoding='utf-8').readlines()
```

```python
백설공주파일
```

```python
백설공주1 = []

for i in range(0,len(백설공주파일)) :
    text = 백설공주파일[i]
    text = re.sub('[^가-힣\s\.]','',text)
    text = text.rstrip()
    text = text.lstrip()
    백설공주1.append(text)
```

```python
백설공주2 = []

for i in range(0,len(백설공주1)):
    if 백설공주1[i] >= ' ':
        백설공주2.append(백설공주1[i]) 
```

```python
백설공주 = []

for line in 백설공주2:
    replace_line = line.replace('\n\r','')
    백설공주.append(replace_line)
```

```python
from konlpy.tag import Mecab
tagger = Mecab(dicpath="C:\\mecab\\mecab-ko-dic")
mecab = Mecab(dicpath="C:\\mecab\\mecab-ko-dic")
```

```python
백설공주pd = pd.DataFrame(백설공주)
```

```python
백설공주pd[0] = 백설공주pd[0].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣]",'')
```

```python
from konlpy.tag import Okt
```

```python
okt = Okt()
```

```python
stopwords = '은 는 이 가 을 를 의 도 또 과 와 즉 좀 등 더'
stopwords = stopwords.split(' ')
```

```python
백설공주_train = []

for word in 백설공주pd[0]:
    temp_X = []
    temp_X = okt.morphs(word, stem=True)
    temp_X = [word for word in temp_X if not word in stopwords]
    백설공주_train.append(temp_X)
```

```python
백설공주님 = []

for i in range(0,len(백설공주_train)):
    for r in range(0,len(백설공주_train[i])):
        백설공주님.append([백설공주_train[i][r]])
```
