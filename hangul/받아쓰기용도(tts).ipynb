```python
df_pd = pd.DataFrame(df)

df_pd[0] = df_pd[0].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣]",'')

from konlpy.tag import Okt

okt = Okt()

stopwords = '은 는 이 가 을 를 의 도 또 과 와 즉 좀 등 더'
stopwords = stopwords.split(' ')

df_train = []

for word in df_pd[0]:
    temp_X = []
    temp_X = okt.morphs(word, stem=True)
    temp_X = [word for word in temp_X if not word in stopwords]
    df_train.append(temp_X)

# df_train
```

```python
df_list = []

for i in range(0,len(df_train)):
    for r in range(0,len(df_train[i])):
        df_list.append([df_train[i][r]])
```

```python
text = []

for i in df_list:
    if len(i[0]) == 3:
        text.append(i)
```

```python
print(text)
```

```python
# decompose
dc= []
for i in range(0, len(text)):
    dc.append(hgtk.text.decompose(text[i][0]))
```

```python
# select
select2="ㄷᴥㅂ"
hard_list= []
for select in dc:
    if select2 in select or "ㅁᴥㄴ" in select :
        hard_list.append(select)
```

```python
print(hard_list)
```

```python
# compose
cm = []
for i in range(0, len(hard_list)):
    cm.append(hgtk.text.compose(hard_list[i]))
print(cm)
```
