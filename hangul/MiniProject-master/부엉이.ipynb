```python
import pandas as pd
import re
```

```python
부엉이1 = []

for i in range(0,len(부엉이)):
    text = 부엉이[i]
    text = re.sub('[^가-힣\s]','',text)
    text = text.rstrip()
    text = text.lstrip() 
    부엉이1.append(text)
```

```python
부엉이1
```

```python
부엉이2 = []

for i in range(0, len(부엉이1)):
    if 부엉이1[i] >= ' ':
        부엉이2.append(부엉이1[i])
```

```python
부엉이2
```

```python
부엉이3 = []

for line in 부엉이2:
    replace_line = line.replace('\n','')
    부엉이3.append(replace_line)
```

```python
부엉이3
```

```python
full_text=' '
for line in 부엉이3:
    full_text=full_text + ''+str(line)+ ' '
full_text_lines = full_text.split("\r\n")
```

```python
stop_words = "은 는 이 가 을 를 있 었 의 도 또"
stop_words=stop_words.split(' ')
stop_words[0:10]
```

```python
nouns = []
for post in 부엉이당:
    for noun in tagger.nouns(post):
        if noun not in stop_words:
            nouns.append(noun)
print(nouns[0:30])
```

```python
부엉이연습['문장'] = 부엉이연습['문장'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","")
```

```python
from konlpy.tag import Okt
```

```python
okt = Okt()
```

```python
stopwords = ['의','가','이','은','들','는','좀','잘','걍','과','도','를','으로','자','에','와','한','하다']
```

```python
X_train = []
for sentence in 부엉이연습['문장']:
    temp_X = []
    temp_X = okt.morphs(sentence, stem=True) # 토큰화
    temp_X = [word for word in temp_X if not word in stopwords] # 불용어 제거
    X_train.append(temp_X)
```
