```python
from bs4 import BeautifulSoup
import requests


headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.121 Safari/537.36'}

url = "https://map.naver.com/v5/entry/place/11571730"
```

```python
response = requests.get(url, headers = headers)
response
```

```python
soup = BeautifulSoup(response.content, "html.parser")   # response.text면 글 깨짐
soup
```

```python
# soup에는 "덕수궁" 없음

names = soup.findAll('span', {'class': '_3XamX'})   #  <- F12에서 "덕수궁"위치
```

```python
print(names)
```
